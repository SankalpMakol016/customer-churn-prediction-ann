```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder,OrdinalEncoder
import pickle
```

```python
df = pd.read_csv("Churn_Modelling.csv")
df
```

```python
df = df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df
```

```python
label_encoder_gender = LabelEncoder()
df['Gender'] = label_encoder_gender.fit_transform(df['Gender'])
df
```

```python
onehot_encode_geography = OneHotEncoder()
geo_encode = onehot_encode_geography.fit_transform(df[['Geography']])
geo_encode
```

```python
onehot_encode_geography.get_feature_names_out(['Geography'])
```

```python
print(geo_encode.toarray())
```

```python
geo_encode_df = pd.DataFrame(geo_encode.toarray(),columns=onehot_encode_geography.get_feature_names_out(['Geography']))
geo_encode_df
```

```python
df=df.drop(columns=onehot_encode_geography.get_feature_names_out(['Geography']))
df
```

```python
df = pd.concat([df,geo_encode_df],axis=1)
df
```

```python
df = df.drop(columns=['Geography'])
```

```python
with open ('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)
    
with open ('onehot_encode_geography.pkl','wb') as file:
    pickle.dump(onehot_encode_geography,file)
```
